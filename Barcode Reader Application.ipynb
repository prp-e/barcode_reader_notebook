# Reading the barcode

```python
import cv2
from pyzbar import pyzbar
import matplotlib.pyplot as plt
```

```python
barcode_image = cv2.imread('barcode.jpg')

plt.imshow(barcode_image)
plt.show()
```

```python
barcodes = pyzbar.decode(barcode_image)
print(barcodes)
```

```python
barcode = barcodes[0]
```

```python
barcode.data
```

```python
x, y, w, h = barcode.rect
```

```python
cv2.rectangle(barcode_image, (x, y), (x + w, y + h), (0,0,255), 5)

plt.imshow(cv2.cvtColor(barcode_image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
print(f'{int(barcode.data)} has been read successfully.')
```
